```python
import cv2
import gym
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from stable_baselines3 import DQN
from environment import AfcEnvGrid
```

```python
model = DQN.load("dqn-fix2")
env = AfcEnvGrid()
```

```python
arr_fixs = []
arr_rwds = []

obs = env.reset()
print(obs)
for i in range(200):
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    
    arr_rwds.append(reward)
    
    # fixs
    scale  = np.array([int(env.an_height/env.im_height), int(env.an_width/env.im_width)])
    fixs = env.fixs * scale
    fixs = [fixs[1], fixs[0]]
    arr_fixs.append(fixs)

    # env.render()
    if done:
        obs = env.reset()

env.close()
```

```python
arr_fixs = np.array(arr_fixs)

img = np.uint8(cv2.imread("./data/images/ADE_train_00010429.jpg"))
plt.imshow(img, "gray")
plt.plot(arr_fixs[:, 0], arr_fixs[:,1], color="red", marker="+")
```

```python
arr_fixs = np.array(arr_fixs)

img = np.uint8(cv2.imread("./data/annotations/ADE_train_00010429.png"))
img = np.where(img == 20, 255, img)
plt.imshow(img, "gray")
plt.plot(arr_fixs[:, 0], arr_fixs[:,1], color="red", marker="*")
```
